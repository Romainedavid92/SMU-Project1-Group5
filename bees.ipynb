```python
#validate data file and group
bees_year_df = bees_one_df.groupby(["State", "Year"]).sum().reset_index()
bees_year_df
```

```python
#group by year, see differences for US
US_bees_year_df = bees_one_df.groupby(["Year"]).sum().reset_index()
US_bees_year_df
```

```python
# find % Lost of total
US_bees_lost = US_bees_year_df["Lost Colonies"]/US_bees_year_df["Colonies"]
US_bees_lost
```

```python
#conccat and create new df
US_Lost_df = pd.concat([US_bees_year_df,US_bees_lost], axis=1)
US_Lost_df.head()
```
